# Create a DF that has all the essential aggregations at customer level

```python
customer_df = number_of_unique_quarters_a_customer_transacted_in.merge(right=total_transactions_per_user, on ='user_profile_id')
customer_df.rename(columns={'count': 'number_of_unique_quarters_they_transacted'}, inplace=True)
customer_df.head()
```

```python
# create bins/buckets based on number of transctions and number of quarters a customer has transacted in
# this will help to visualise and isolate one off vs repeat customers
customer_df['transaction_bin']= pd.cut(customer_df['transaction_count'], bins=[0, 1, 2 ,5 ,10, 100, 1000, 5000, 10000, 100000, float('inf')])
customer_df['quarter_bin']= pd.cut(customer_df['number_of_unique_quarters_they_transacted'], bins=[0,1,2,4,8,16,32,64])
```

```python
# pivot data to prepare for heatmap visualisation later on
pivot = customer_df.reset_index().pivot_table(
    index = 'transaction_bin',
    columns = 'quarter_bin',
    values = 'user_profile_id',
    aggfunc = 'nunique',
    fill_value = 0
)
pivot
```

# Define customer segment based on findings

```python
def classify_customer(row):
    if row['transaction_count']==1 and row['number_of_unique_quarters_they_transacted'] == 1:
        return 'One Off'
    elif row['transaction_count']<=10 and row['number_of_unique_quarters_they_transacted'] <= 2:
        return 'Low Engagement'
    else:
        return 'Repeat'
customer_df['customer_frequency'] = customer_df.apply(classify_customer, axis=1)      
```

```python
customer_df.reset_index(inplace=True)
number_of_customers_per_segment = customer_df.groupby('customer_frequency')['user_profile_id'].nunique()
number_of_customers_per_segment = number_of_customers_per_segment.reset_index()
number_of_customers_per_segment.rename(columns={'user_profile_id':'customer_count'}, inplace=True)
```

# Questions
- Is the overall volume and revenue driven mostly by repeats?
- Are low engagement customers growing or dropping off?

```python
# Merge customer frequency info into all_transactions_df
all_transactions_df = all_transactions_df.merge(
    customer_df[['user_profile_id', 'customer_frequency']],
    on='user_profile_id',
    how='left'
)
time_metrics = all_transactions_df.groupby(
    ['quarter_created_date', 'customer_frequency']
).agg({
    'transaction_count': 'sum',
    'customer_volume_gbp': 'sum',
    'customer_revenue_gbp': 'sum',
    'user_profile_id': 'nunique'
}).reset_index().rename(columns={'user_profile_id':'customer_count'})
fig, axs = plt.subplots(1, 2, figsize=(20, 5), sharey=False)

sns.lineplot(data=time_metrics, x='quarter_created_date', y='customer_revenue_gbp', hue='customer_frequency', ax=axs[0])
axs[0].set_title('Revenue Over Time by Customer Frequency')
sns.lineplot(data=time_metrics, x='quarter_created_date', y='customer_volume_gbp', hue='customer_frequency', ax=axs[1])
axs[1].set_title('Volume Over Time by Customer Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create pivot table for stacked data
stacked_data = time_metrics.pivot_table(
    index='quarter_created_date', 
    columns='customer_frequency', 
    values='customer_revenue_gbp',
    aggfunc='sum'
)
# Plot stacked bar chart
ax = stacked_data.plot(kind='bar', stacked=True, figsize=(10, 6))

# Format y-axis in billions
import matplotlib.ticker as ticker

def billions(x, pos):
    return f'{x/1000000:.1f}m'

# Apply the formatter to the y-axis
ax.yaxis.set_major_formatter(ticker.FuncFormatter(billions))

# Format x-axis labels to show quarters nicely
quarter_labels = []
for date in stacked_data.index:
    # Extract year and month from the datetime
    year = date.year
    month = date.month
    # Determine quarter
    quarter = (month - 1) // 3 + 1
    quarter_labels.append(f"Q{quarter} {year}")

# Apply the formatted labels
ax.set_xticklabels(quarter_labels, rotation=45)

plt.title('Revenue Over Time by Customer Frequency')
plt.xlabel('Quarter')
plt.ylabel('Customer Revenue (GBP)')
plt.legend(title='Customer Frequency')
plt.tight_layout()
plt.show()
```

YoY growth Per Customer Frequency

```python
time_metrics['revenue_growth'] = time_metrics.groupby('customer_frequency')['customer_revenue_gbp'].pct_change()
time_metrics = time_metrics[(time_metrics['quarter_created_date'] >= '2020-01-01') & (time_metrics['quarter_created_date'] < '2024-04-01')]
time_metrics = time_metrics.sort_values(['customer_frequency', 'quarter_created_date'])

# Calculate YoY growth
time_metrics['revenue_yoy_growth'] = time_metrics.groupby('customer_frequency')['customer_revenue_gbp'].pct_change(periods=4)

# Smooth with rolling average (2-quarter smoothing)
time_metrics['revenue_yoy_smoothed'] = time_metrics.groupby('customer_frequency')['revenue_yoy_growth'].transform(lambda x: x.rolling(window=2, min_periods=1).mean())

# Smoothed YoY growth
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=time_metrics,
    x='quarter_created_date',
    y='revenue_yoy_smoothed',
    hue='customer_frequency',
    marker='o'
)

plt.title('Smoothed YoY Revenue Growth by Customer Segment')
plt.xlabel('Quarter')
plt.ylabel('Smoothed Revenue Growth (YoY %)')
plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(title='Customer Segment')
plt.show()
```

```python
# Calculate share of customers and revenue by segment
summary = customer_df.groupby('customer_frequency').agg({
    'user_profile_id': 'nunique',
    'customer_revenue_gbp': 'sum'
}).rename(columns={'user_profile_id': 'customer_count'})

# Add total calculations
total_customers = summary['customer_count'].sum()
total_revenue = summary['customer_revenue_gbp'].sum()

# Calculate % share columns
summary['customer_pct'] = (summary['customer_count'] / total_customers * 100).round(2)
summary['revenue_pct'] = (summary['customer_revenue_gbp'] / total_revenue * 100).round(2)

summary_formatted = summary.copy()
summary_formatted['customer_revenue_gbp'] = summary_formatted['customer_revenue_gbp'].apply(lambda x: f"{x:,.0f}")
summary_formatted
```

# Main Findings and Conclusion

- Repeat customers are the main revenue drivers  of the business. They represent just over half of all customers (~55%) and they generate ~94% of total revenue validating a strong focus on retention and engagement strategies.
- On the other hand the One-Off and Low Engagement users collectively account for nearly 45% of customers but contribute minimal revenue. - These groups are largely transactional and short-lived.
Engagement duration matters: Bins/Brackets with users transacting over multiple quarters  especially 10100 transactions across 816 quarters  generate the most revenue. Sustained activity is a key indicator of customer value.
- The most valuable segments are not just active, but consistent over time, underscoring the value of long-term engagement rather than acquisition volume alone.

# Suggested Next Steps:

- Refine segmentation further: Deepen the current model by splitting the Repeat segment into sub-cohorts (e.g., high-frequency-high-revenue customers vs. high- frequency-low-revenue) to uncover more targeted opportunities.
- Targeted retention strategies: Develop initiatives to nurture Low Engagement users toward Repeat behavior.
- Explore needs of high-engagement users. Look into what else frequent users need  such as multi-currency features, business use cases, or automation tools  to deepen their relationship with Wise.
- Behavioral and revenue forecasting: Use this binning framework to predict future revenue contributions by customer segment, aiding long-term planning.
